## DATA VISUALIZATION

```python
# Importing module from python library

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Distribution of Age using Histogram

plt.figure(figsize=(10,6))    # figure size
plt.hist(df['Age'], bins=10,edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')        #no.of patients
plt.show()
```

```python
# Gender Distribution using Pie chart

gender_counts = df['Gender'].value_counts()

plt.figure(figsize=(6,6))  #figure size
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%',colors=['lightcoral', 'lightskyblue'])
plt.title('Gender Distribution')
plt.show()
```

```python
# Heart Disease Distribution using count plot

plt.figure(figsize=(8,6))
sns.countplot(x='Heart_Disease', data=df,palette=['orange','red'])
plt.title('Heart Disease Distribution')
plt.show()
```

```python
# Correlation Heatmap (Numerical Data)

numerical_df = df.select_dtypes(include=['number'])  # Select columns with numerical data types

plt.figure(figsize=(8,5))
sns.heatmap(numerical_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()

#Positive correlations are shown in shades of red.
#Negative correlations are shown in shades of blue.
#Values close to 1 indicate a strong positive correlation.
#Values close to -1 indicate a strong negative correlation.
```

```python
# Cholesterol Levels vs. Heart Disease using boxplot

plt.figure(figsize=(10,6))
sns.boxplot(x='Heart_Disease', y='Cholesterol_Level (mg/dL)', data=df)
plt.title('Cholesterol Levels by Heart Disease')
plt.xlabel('Heart Disease')
plt.ylabel('Cholesterol Level (mg/dL)')
plt.show()

#No outliers
```

```python
# Scatter Plot
plt.figure(figsize=(10,6))  # Set figure size
plt.scatter(df['Age'], df['Cholesterol_Level (mg/dL)'], color='blue')

# Add labels and title
plt.title('Scatter Plot: Age vs. Cholesterol Level')
plt.xlabel('Age')
plt.ylabel('Cholesterol Level (mg/dL)')

# Show the plot
plt.show()
```

```python
# Exercise Hours per Week Distribution

plt.figure(figsize=(10, 6))                #fig size
sns.histplot(df['Exercise_Hours_Per_Week'], bins=10, color='green')
plt.title('Exercise Hours Per Week Distribution')
plt.xlabel('Hours')
plt.ylabel('count')
plt.show()
```
